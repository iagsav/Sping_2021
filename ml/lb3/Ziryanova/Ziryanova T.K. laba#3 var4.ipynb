#### 	8. Решите поставленную задачу классификации в соответствии с заданием. При подборе параметров классификатора используйте метод GridSearchCV и перекрёстную проверку 

##### 1 Классификатор K ближайших соседей (задаётся количество ближайших объектов)

```python
Y = A[:,0]
X = A[:,1:]
# класс гриба запишем в переменную Y, остальные признаки - в X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1000, random_state=5)
# разделим датасет на обучающую и контрольную выборки
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()
print(classifier.get_params())
# выбрали классификатор и выгрузили его стандартные параметры
```

```python
parameters = {
    "p": [1, 2],
    "weights": ["uniform", "distance"],
    "n_neighbors": [1, 2, 3, 4, 5, 6, 7, 8, 9]
}
# выбираем параметры, изменения которых хотим исследовать
```

```python
gridSearch = GridSearchCV(classifier, parameters, n_jobs=4)
# Создаем GridSearchCV. Передаем наш классификатор и его параметры
gridSearch.fit(X_train, Y_train)
# Запуск GridSearchCV. Для каждого значения параметра из 'parameters' он запустит классификатор на тестовом датасете
# признаков (X_train) и сравнит ответ классификатора с настоящим (Y_train)
```

```python
print(gridSearch.best_params_)
# выводим наилучший параметр
```

```python
cvResults = pd.DataFrame(gridSearch.cv_results_)
cvResults[["param_p", "param_n_neighbors", "param_weights", "mean_test_score"]]
# полный отчет по параметрам
# наилучший результат мы получаем тогда, когда колличество ближайших объектор равно единице при равномерном весе 
# наихудший результат мы получаем при количестве объектов больше 8 при равномерном весе (его максимум равен 0,999)
# при весе точек, обратному расстоянию, значения столбца mean_test_score принимают значение, равное единице при любом весе
```

##### 2 Логический классификатор (бинарное решающее дерево).

```python
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
# выбор классификатора
```

```python
print(treeClassifier.get_params())
# вывод параметров по умолчанию
```

```python
parameters = {
    "max_depth": [3, 6, 9, 12, 15], # Максимальная глубина дерева. Если нет, то узлы расширяются до тех пор, пока все листья не станут чистыми 
    "max_features": [50, 90, 110], # Количество функций, которые следует учитывать при поиске лучшего разделения
    "min_samples_leaf": [3, 5, 10, 90] # Минимальное количество выборок, необходимых для нахождения в листовом узле.
}
# выбор параметров для исследования
```

```python
gridSearchTree = GridSearchCV(treeClassifier, parameters, n_jobs=4)
# Создаем GridSearchCV. Передаем наш классификатор и его параметры
gridSearchTree.fit(X_train, Y_train)
# Запуск GridSearchCV. Для каждого значения параметра из 'parameters' он запустит классификатор на тестовом датасете
# признаков (X_train) и сравнит ответ классификатора с настоящим (Y_train)
```

```python
print(gridSearchTree.best_params_)
# выводим наилучший параметр
```

```python
cvResults = pd.DataFrame(gridSearchTree.cv_results_)
cvResults[["param_max_depth", "param_max_features", "param_min_samples_leaf", "mean_test_score"]]
# полный отчет по параметрам
```

Задача бинарной классификации гриба на типы "съедобный"/"несъедобный" успешно решена с использованием двух типов классификаторов: KNeighborsClassifier и DecisionTreeClassifier.
При правильном подборе параметров каждого классификатора удалось добиться значений метрик precision и recall по 100%, что говорит об абсолютной точности работы классификатора на тестовой выборке.
